```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from datetime import date, datetime
import csv
import time
```

```python
class amlo_scraping(object):

    def __init__(self, n_sections):
        self.n_sections =  n_sections
        self.speech_id = 0
        self.url = 'https://lopezobrador.org.mx/page/'
        self.speechs = []
        self.id = time.time_ns()
        self.filename =  f'amlo-speechs-{self.id}.csv'

    def __getMonth(self, m):
        months = {'enero':'1', 'febrero':'2',
              'marzo':'3', 'abril':'4',
              'mayo':'5', 'junio':'6',
              'julio':'7','agosto':'8',
              'septiembre':'9','octubre':'10',
              'noviembre':'11','diciembre':'12'}
        return months[m]
    
    def __preprocess_date(self, d):
        f = d.split(' ')
        date = datetime(year=int(f[2]), month=int(self.__getMonth(f[0])), day=int(f[1].replace(',', '')))
        return date               
    
    def run_scrapping(self):
        for i in range(1, self.n_sections + 1):
            print("processing: ", i)
            self.speech_id += i
            url = self.url + str(i) + '/'
            page = requests.get(url)
            soup = BeautifulSoup(page.content, 'lxml')
            articles = soup.find_all('div', class_='row isotope-container')
            for s in articles:
                for element in s.findChildren('article'):
                    title = element.find_all('h2',class_='entry-title')
                    date = element.find_all('span',class_='entry-date')
                    dte = self.__preprocess_date(date[0].text) 
                    uinternal_url = title[0].find_all('a',href=True)[0]['href']       
                    npage = requests.get(uinternal_url)
                    nsoup = BeautifulSoup(npage.content, 'html.parser')
                    content = nsoup.find_all('div', class_='entry-content')
                    if content[0].em is not None:
                        content[0].em.clear()
                    dct = {'id_speech': self.speech_id, 'date': dte, 'title': title[0].text , 'url': uinternal_url, 'content': content[0].text}
                    self.speechs.append(dct)
                    self.speech_id += i

    def save_data(self):
        print("saving speechs")
        keys = self.speechs[0].keys()
        with open(self.filename, 'w', encoding='utf-8', newline='')  as output_file:
            dict_writer = csv.DictWriter(output_file, keys)
            dict_writer.writeheader()
            dict_writer.writerows(self.speechs)
        print('file saved')
```

```python
ds = amlo_scraping(666)
ds.run_scrapping()
ds.save_data()
```
